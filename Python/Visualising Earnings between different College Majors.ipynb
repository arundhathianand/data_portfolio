## Most Popular Majors

```python
recent_grads.head()
```

```python
df = recent_grads[['Major','Total','Median']]
df2 = df.sort_values('Total',ascending=False)
df2

```

```python
df3=df2[['Major','Total', 'Median']]
df3.drop('Major', inplace=True, axis=1)
```

```python
df3=(df3-df3.mean())/df3.std()
df3
```

```python
fig = plt.figure(figsize=(30,10))
ax1 = plt.plot(recent_grads['Major'], df3['Total'], c='red')
ax2 = plt.plot(recent_grads['Major'], df3['Median'], c='blue')
```

## Histograms

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(5,12))
for r in range(0,4):
    ax = fig.add_subplot(4,1,r+1)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=30)
```

```python

cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(5,12))
for r in range(4,8):
    ax = fig.add_subplot(4,1,r-3)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=30)
```
